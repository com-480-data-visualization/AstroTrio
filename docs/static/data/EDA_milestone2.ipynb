## 2 - Solar system
### Solar system major bodies data

```python
# Download dataset
path = kagglehub.dataset_download("jaredsavage/solar-system-major-bodies-data")

print("Path to dataset files:", path)
!ls -lh $path
```

```python
df_solar_system = pd.read_csv(path + "/sol_data.csv")
df_solar_system.shape
```

```python
df_solar_system.head()
```

```python
df_solar_system.columns
```

```python
df_solar_system.nunique()
```

```python
# First : focus on the planets
df_planets = df_solar_system[df_solar_system['isPlanet'] == True]
df_planets.shape
```

```python
df_planets
```

```python
# Create visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of planets vs non-planets
plt.figure(figsize=(8, 6))
planet_counts = df_solar_system['isPlanet'].value_counts()
plt.pie(planet_counts, labels=['Non-Planets', 'Planets'], autopct='%1.1f%%', 
        colors=['#ff9999','#66b3ff'], startangle=90)
plt.title('Distribution of Planets vs Non-Planets in the Solar System')
plt.axis('equal')
plt.show()

# 2. Scatter plot of mass vs semimajor axis for planets
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_planets, x='semimajorAxis_AU', y='massj', 
                size='gravity', hue='eName', sizes=(100, 1000))
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Semimajor Axis (AU)')
plt.ylabel('Mass (Jupiter Masses)')
plt.title('Planet Mass vs Distance from Sun')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. Bar plot of planet densities
plt.figure(figsize=(12, 6))
sns.barplot(data=df_planets, x='eName', y='density')
plt.xticks(rotation=45)
plt.xlabel('Planet')
plt.ylabel('Density (g/cm³)')
plt.title('Density of Planets in the Solar System')
plt.tight_layout()
plt.show()

# 4. Heatmap of correlations between numerical features
numerical_features = ['semimajorAxis_AU', 'massj', 'density', 'gravity', 'eccentricity', 'inclination']
plt.figure(figsize=(10, 8))
sns.heatmap(df_planets[numerical_features].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Planet Features')
plt.tight_layout()
plt.show()
```

```python
# Do a radar graph for planet = 'Mars'
# Select Mars data
mars_data = df_solar_system_2[df_solar_system_2['Planet'] == 'Mars']

# Get the numerical features for Mars
mars_values = mars_data[columns_radar].values.flatten()

# Set up the angles for the radar chart (divide the plot / number of variables)
angles = np.linspace(0, 2*np.pi, len(columns_radar), endpoint=False)

# Close the plot by appending the first value
values = np.concatenate((mars_values,[mars_values[0]]))
angles = np.concatenate((angles,[angles[0]]))

# Initialize the spider plot
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, polar=True)

# Plot data
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)

# Set the labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(columns_radar, size=8)

# Add title
plt.title("Mars Features Radar Chart")

# Save the figure
plt.savefig("../images/mars_radar_chart.png", dpi=300)

plt.tight_layout()
plt.show()

```

```python
# Craft the feature Main Composition --> first element from the Composition column
df_solar_system_2['Main Composition'] = df_solar_system_2['Composition'].str.split(' ').str[0].str.split(',').str[0]
df_solar_system_2['Main Composition']
```

```python
# Create temperature range visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to extract min and max temperatures from range string
def extract_temp_range(temp_str):
    if pd.isna(temp_str) or temp_str == '':
        return pd.NA, pd.NA
    try:
        # Handle single temperature case
        if ' to ' not in str(temp_str):
            temp = float(temp_str)
            return temp, temp
        # Handle range case
        min_temp, max_temp = map(float, temp_str.split(' to '))
        return min_temp, max_temp
    except:
        return pd.NA, pd.NA

# Extract temperature ranges
temp_data = pd.DataFrame()
temp_data['Planet'] = df_solar_system_2['Planet']
temp_data[['min_temp', 'max_temp']] = df_solar_system_2['Surface Temperature (C)'].apply(extract_temp_range).apply(pd.Series)

# Sort by average temperature
temp_data['avg_temp'] = (temp_data['min_temp'] + temp_data['max_temp']) / 2
temp_data = temp_data.sort_values('avg_temp', ascending=True)

# Create the visualization
plt.figure(figsize=(8, 8))

# Plot temperature ranges
for i, row in temp_data.iterrows():
    if not pd.isna(row['min_temp']) and not pd.isna(row['max_temp']):
        plt.hlines(y=row['Planet'], xmin=row['min_temp'], xmax=row['max_temp'], 
                  color='skyblue', alpha=0.7, linewidth=10)
        
        # Add points for min and max temperatures
        plt.plot(row['min_temp'], row['Planet'], 'o', color='navy', markersize=8)
        plt.plot(row['max_temp'], row['Planet'], 'o', color='red', markersize=8)

# Customize the plot
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)  # Add line at 0°C
plt.grid(True, alpha=0.3)
plt.title('Surface Temperature Ranges of Solar System Bodies', pad=20)
plt.xlabel('Temperature (°C)')
plt.ylabel('Celestial Body')

plt.tight_layout()
plt.show()
```
